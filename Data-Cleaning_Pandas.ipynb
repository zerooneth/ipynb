# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Misc2\Work in progress\Others\IT\Learning\Python\Pandas\Customer Call List.xlsx")
df
```

```python
# usually the first thing. punem df = ca as suprascriem si sa salvam modificarile. Am putea si cu inplace = True

df = df.drop_duplicates()
df
```

```python
# removing columns that we don't need

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# Curatam coloana Last_Name. Cu strip poti sterge doar din laterale.
# nu punem df = pt ca ne-ar distruge tot df. aplicam doar pe coloana aia.

df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df

# sau putem sa facem asa:
# df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
# cleaning phone number ca sa arate ca 123-545-5421
# get rid of non numeric values. strip it down to just the number. 
# [^a-zA-Z0-9] - return any character except a-z A-Z 0-9
# then reformatting

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
# putem folosi for loop sau lambda. aici folosim lambda
# trebuie sa convertim prima data in strings

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# get rid of nan --, Na--

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# clean address, put it in 3 sepparate columns. zip, street, city
# split the columns. First value - the splitting value. 2nd - how many values from left to right
# n = additional columns

df['Address'].str.split(',', n=2, expand = True)
```

```python
# in stanga punem adresele pe care le vrem
# they will be appended, don't replace the actual address column

df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand = True)
df
```

```python
# Clean Paying Customer si Do_Not_Contact

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('No','N')
df
```

```python
# stergem valorile NA din tot DF-ul
# dar ca sa putem, ele nu sunt string, ele spun ca nu este number. Asa ca trebuie sa facem fillna

df = df.fillna('')

# Apoi, unde e chiar string NA facem ca mai jos
df = df.replace('N/a','')
df = df.replace('N/a','NaN')
df
```

```python
# get a list for people that they can call. so sa fie N la do_not_contact si...

for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace=True)

df
```

```python
# si sa existe phone number

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x,inplace=True)

df

# puteam sa facem si drop values
# df.dropna(subset='Phone_Number', inplace=True)
```

```python
# resetam indexul pt df ca sa arate bine in varianta finala

df = df.reset_index(drop=True)
df
```
