# Exploratory Data Analysis

The first look at you data. Identifying patterns within the data, understanding the relationships between the features, looking at outliers that may exist within your data set. Patterns, mistakes, missing values, things to clean up during cleaning process

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Misc2\Work in progress\Others\IT\Learning\Python\Pandas\world_population2.csv")
df
```

```python
# change decimal points

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# all the collumns and how many missing values
# putem sa o populam cu median value, putem sa stergem complet acele tari

df.isnull().mean()
```

```python
# how many unique values

df.nunique()
```

```python
# top range per one column

df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
# looking at correlations, between usually numeric values
# method='pearson', min_periods=1, numeric_only=True / {‘pearson’, ‘kendall’, ‘spearman’}

df.corr(numeric_only=True)
```

```python
# visualize this using a heatmap

sns.heatmap(df.corr(numeric_only=True), annot = True)

# il facem mai mare
plt.rcParams['figure.figsize'] = (10,7)

plt.show()
```

```python
# Group together data. Look at the continent. Are there certain continents that have grown faster than others and in which way?
# atentie. nu a mers mean pana acum pt ca sunt valori non numerice. trebuie sa trecem numeric_only=True

df.groupby(by='Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
# prin [] specificam doar coloanele pe care le vrem

df2 = df.groupby(by='Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)

# transpose este sa schimbe randurile cu coloanele
df3 = df2.transpose()

df2.plot()
```
