# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Misc2\Work in progress\Others\IT\Learning\Python\Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python

# tipuri: kind = 'line' default /  subplots = True - face 3 ploturi separate 

df.plot(kind = 'line', title = "Ice Cream Ratings", xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# normal sau stacked. putem specifica coloana exacta df['Flavor Rating'].plot

df.plot(kind = 'bar', stacked = True)
```

```python
# horizontal 

df.plot.barh(stacked = True)
```

```python
# Scatter. Trebuie sa specificam neaparat x si y axis. S = size. C = color

df.plot.scatter(x='Texture Rating', y='Overall Rating', s = 300, c = 'Red')
```

```python
# Histogram. For showing distribution of variables

df.plot.hist(bins = 20)
```

```python
# 1 minimum value within that column. 
# 2 bottop of box 25% percentile of all values 
# 3 green line 50 percent
# 4 75 percent top of the box
# 5 maximul value

df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize = (10,6))
```
